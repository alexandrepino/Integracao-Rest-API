# Importando Bibliotecas utilizadas no projeto

```python
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt
```

# Trasformando a URL da API em uma constante

```python
URL_MOEDAS = "https://economia.awesomeapi.com.br/json/all"
```

# Criando a função que puxa os dados da API

```python
def requests_moedas():
    r = requests.get(URL_MOEDAS)
    r_json = r.json()
    return r_json
```

```python
cotacao_moedas = requests_moedas()
```

# Exibindo os dados puxados da API em um dicionário

```python
cotacao_moedas
```

# Transformando o dicionário com os dados da API em um DataFrame

```python
cotacao_moedas_df = pd.DataFrame(cotacao_moedas).transpose()
```

# Exibindo a tabela do DataFrame na tela

```python
cotacao_moedas_df
```

# Exibindo o valor de compra das moedas

```python
print('O valor atual do Dólar Americano é R${:.2f}'.format(float(cotacao_moedas_df['bid']['USD'])))
print('O valor atual do Dólar Canadense é R${:.2f}'.format(float(cotacao_moedas_df['bid']['CAD'])))
print('O valor atual do Bitcoin é R${:,}'.format(float(cotacao_moedas_df['bid']['BTC'])))
print('O valor atual do Libra é R${:.2f}'.format(float(cotacao_moedas_df['bid']['GBP'])))
print('O valor atual do Peso Argentino é R${:.2f}'.format(float(cotacao_moedas_df['bid']['ARS'])))
print('O valor atual do EURO é R${:.2f}'.format(float(cotacao_moedas_df['bid']['EUR'])))
print('O valor atual do Fraco Suíço Canadense é R${:.2f}'.format(float(cotacao_moedas_df['bid']['CHF'])))
print('O valor atual do Yuan Chinês é R${:.2f}'.format(float(cotacao_moedas_df['bid']['CNY'])))
print('O valor atual do Novo Shekel Israelense é R${:.2f}'.format(float(cotacao_moedas_df['bid']['ILS'])))
```
